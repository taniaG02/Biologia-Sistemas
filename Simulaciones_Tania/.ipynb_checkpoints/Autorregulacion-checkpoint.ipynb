##### Diagrama de fases --> ver puntos en el equilibrio y estabilidad.

```python
## suponemos que el volumen celular es constante e igual a 1 
## cogemos la ecuaciones y lo igualamos a 0
## dibujamos el diagrama de bifurcación.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros (valores obtenidos de Isaacs et al. 2003)
m = 60        # Número de copias del plásmido
alpha = 11    # Activación cooperativa
c1 = 0.05     # Constante de equilibrio dimerización cI
c2 = 0.33     # Constante de equilibrio cI–OR1
s1 = 2.0      # Afinidad relativa OR2 vs OR1
s2 = 0.08     # Afinidad relativa OR3 vs OR1
beta = 100.0  # Producción basal de cI por ciclo celular
gamma_g = 10.0  # Degradación de GFP (no usado aquí)
eta = 0.01    # Eficiencia relativa de GFP (no usado aquí)
v = 1       # Volumen celular adimensional
```

```python
# Funciones f(x,v), h(x,v) y dx/dt
def d0(x, v):
    return m / (1 + c1*c2*x**2/v**2 + s1*(c1*c2)**2*x**4/v**4 + s1*s2*(c1*c2)**3*x**6/v**6)

def f(x, v):
    return m * (1 + (c1*c2)*x**2/v**2 + alpha*s1*(c1*c2)**2*x**4/v**4) / \
           (1 + (c1*c2)*x**2/v**2 + s1*(c1*c2)**2*x**4/v**4 + s1*s2*(c1*c2)**3*x**6/v**6)

def h(x, v):
    return 1 + 4*c1*x/v + 4*(c1*c2)*d0(x, v)/v**2 + \
           16*s1*(c1*c2)**2*x**3*d0(x, v)/v**4 + 36*s1*s2*(c1*c2)**3*x**5*d0(x, v)/v**6

def dxdt(x, gamma_x):
    return (beta * f(x, v) - gamma_x * x) / h(x, v)
```

```python
# 1) Diagrama de fase: dx/dt vs x para distintos gamma_x
x_vals = np.linspace(0, 2000, 500)
# Configurar la figura una sola vez
plt.figure()

# Iterar sobre los valores de gamma y agregar cada línea a la misma gráfica
for gamma in [500, 1500, 2500, 3750, 3800, 5000, 15000]:
    plt.plot(x_vals, dxdt(x_vals, gamma), label=f'γₓ = {gamma}')

# Elementos comunes
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # Línea horizontal en y=0
plt.title("Diagrama de fase: dx/dt vs x para diferentes γₓ")
plt.xlabel("x (número de monómeros cI)")
plt.ylabel("dx/dt")
plt.xlim(0, 50)
plt.ylim(-2000, 2000)
plt.legend()  # Muestra la leyenda con las etiquetas de cada gamma
plt.grid(True, linestyle=':', alpha=0.7)  # Opcional: agregar grid
plt.show()
```

#### ¿Qué implica la simplificación que hemos hecho de poner volumen constante a 1?
En el modelo original, el volumen celular (v) varía exponencialmente con el tiempo y se divide en cada ciclo celular. Esto afecta las concentraciones de proteínas debido a la dilución durante el crecimiento. Al fijar v=1, simplifico el modelo eliminando la dinámica del volumen, pero los parámetros (como la tasa de degradación) se ajustan para incorporar efectos efectivos de dilución y crecimiento. Asi, ya se incluye no solo la degradación proteica, sino también la "pérdida" de moléculas por división celular. Esto permite que los resultados cualitativos (como la existencia de estados estables y bistabilidad) coincidan con los del artículo, aunque la descripción cuantitativa exacta pueda diferir.

### Cómo varia la concrentación de cI con respecto del tiempo

```python
# Configuración general
dt = 0.001
t = np.arange(0, 15, dt)

# Valores a explorar
gammas = [1000, 2000, 3750, 3800, 5000, 15000]
condiciones_iniciales = [5, 10, 20, 50, 100, 150, 200]

# Crear figura con subplots
fig, axs = plt.subplots(len(gammas), 1, figsize=(10, 15), sharex=True)

# Paleta de colores ampliada (7 colores para 7 condiciones iniciales)
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

# Simulación para cada combinación
for idx, gamma in enumerate(gammas):
    ax = axs[idx]
    
    for color_idx, x0 in enumerate(condiciones_iniciales):
        # Inicialización
        x = np.zeros_like(t)
        x[0] = x0
        
        # Integración numérica
        for i in range(1, len(t)):
            x[i] = x[i-1] + dxdt(x[i-1], gamma) * dt
        
        # Graficar (color correspondiente a cada condición inicial)
        ax.plot(t, x, 
                color=colores[color_idx],
                linestyle='-',
                label=f'x₀ = {x0}' if idx == 0 else "")

    # Configuración de cada subplot
    ax.set_title(f"γₓ = {gamma}", pad=8)
    ax.set_ylabel("cI", fontsize=9)
    ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 50)
    ax.set_xlim(0, 0.25)  # Límites consistentes
    
    # Añadir leyenda solo en el primer subplot
    if idx == 0:
        ax.legend(loc='upper right', title="Cond. Inicial", fontsize=8)

# Configuración común final
plt.xlabel("Tiempo (ciclos celulares adimensionales)", fontsize=10)
plt.tight_layout()
plt.show()
```

### Oscilaciones a altas tasas de degradación
Cuando la tasa de degradación es muy alta, la degradación de cI domina sobre su producción. Esto lleva al sistema a un estado de baja concentración (x≈0).

Sin embargo, la ecuación dx/dt incluye términos no lineales (debido a la retroalimentación positiva y la unión cooperativa al ADN). A altas tasa de degradación, pequeños incrementos en x pueden activar temporalmente la producción (f(x,v)), creando pulsos de producción seguidos de una rápida degradación. Esto genera oscilaciones transitorias.

**Artefactos numéricos:** El método de integración (Euler explícito con dt=0.001) puede amplificar inestabilidades en sistemas rígidos (donde las escalas de tiempo de producción/degradación son muy diferentes). En el modelo original, la división celular actúa como un "reseteo" periódico, amortiguando oscilaciones, pero al omitir este mecanismo, las oscilaciones persisten.

## ¿El modelo determinista presenta histéresis?
Sí, el modelo determinista original predice histéresis. Esto significa que, teóricamente, al aumentar la temperatura (que incrementa la degradación del represor, γₓ), el sistema pasaría de un estado de alta expresión a uno bajo. Si luego se disminuye la temperatura, el modelo sugiere que el sistema no volvería inmediatamente al estado alto, sino que persistiría en el bajo por un tiempo, creando una "memoria" de su estado anterior (histéresis).

```python
# Funciones f(x,v), h(x,v) y dx/dt
def f(x, v):
    return m * (1 + (c1*c2)*x**2/v**2 + alpha*s1*(c1*c2)**2*x**4/v**4) / \
           (1 + (c1*c2)*x**2/v**2 + s1*(c1*c2)**2*x**4/v**4 + s1*s2*(c1*c2)**3*x**6/v**6)

def h(x, v):
    return 1 + 4*c1*x/v + 4*(c1*c2)*d0(x, v)/v**2 + \
           16*s1*(c1*c2)**2*x**3*d0(x, v)/v**4 + 36*s1*s2*(c1*c2)**3*x**5*d0(x, v)/v**6

def dxdt(x, gamma_x):
    return (beta * f(x, v) - gamma_x * x) / h(x, v)

# Derivada numérica de f respecto a x
def dfdx(x, v):
    h = 1e-6
    return (f(x + h, v) - f(x, v)) / h

# Ecuación determinista (dx/dt = 0 en estado estable)
def estado_estable(γx):
    return (beta * f(x) - γx * x) / h(x)

# Rango de γx (degradación) para simular
γx_values_asc = np.linspace(500, 15000, 100)  # Aumentando γx
γx_values_desc = np.linspace(15000, 500, 100) # Disminuyendo γx

def calcular_estados(γx_values, v):
    x_estados = []
    for γx in γx_values:
        x_high = 1000  # Estado alto
        x_low = 1      # Estado bajo
        for _ in range(100):
            x_high = x_high - (beta*f(x_high, v) - γx*x_high) / (beta*dfdx(x_high, v) - γx)
            x_low = x_low - (beta*f(x_low, v) - γx*x_low) / (beta*dfdx(x_low, v) - γx)
        x_estados.append(x_high if x_high > 100 else x_low)
    return x_estados

# Simular
x_asc = calcular_estados(γx_values_asc, v)
x_desc = calcular_estados(γx_values_desc, v)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(γx_values_asc, x_asc, 'r-', label='γx aumentando (Alto → Bajo)')
plt.plot(γx_values_desc, x_desc, 'b--', label='γx disminuyendo (Bajo → Alto)')
plt.xlabel('Tasa de degradación (γx)', fontsize=12)
plt.ylabel('Concentración de represor cI', fontsize=12)
plt.title('Histéresis en el modelo determinista', fontsize=14)
plt.annotate('Transición brusca\n al superar γx crítico', 
             xy=(8000, 800), xytext=(10000, 1500),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
